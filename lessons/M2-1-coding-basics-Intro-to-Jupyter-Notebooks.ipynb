## 1. What is a Jupyter Notebook?

A Jupyter Notebook is an interactive document made of **cells**. The two cell types we will use most are:

- **Markdown cells** for explanations, headings, links, lists, formulas, and interpretation.
- **Code cells** for Python commands that can be executed.

For environmental data analysis, notebooks are useful because they let you keep code, outputs, figures, and interpretation close together. That makes your work easier to inspect, revise, and share.

> A useful mental model: a notebook is a transparent analysis trail, not just a place to run code.

## 5. Variables persist in the kernel

When you run a code cell, Python stores objects such as numbers, text, lists, and tables in memory. These stored objects are called **variables**.

The important notebook workflow idea is this:

> Code cells are separate on the page, but they share the same active Python session while the kernel is running.

Run the next cell. It uses the variable `annual_precip_mm`, which was created in the previous code cell.


````python
annual_precip_inches = annual_precip_mm / 25.4
annual_precip_inches
````

### Kernel-state caution

Because notebooks remember variables, code may work only because you ran an earlier cell. This is useful, but it can also make notebooks confusing.

Good habit: when you think a notebook is finished, use **Restart Kernel and Run All Cells** to check that it runs correctly from top to bottom.


## 6. What is the Python kernel?

The **kernel** is the running Python process that executes your code. Your notebook is the document; the kernel is the engine doing the computation.

The kernel determines:

- Which Python installation runs your code.
- Which packages are available to import.
- Which variables currently exist in memory.
- Whether previous cell outputs reflect the current session.

In this course, you should usually select the kernel associated with your **Miniforge/conda course environment**.

## 8. A tiny environmental data example

The next few cells create and inspect a small dataset. Do not worry about memorizing the code yet. Focus on how notebooks support incremental exploration.


````python
# A small dataset represented with built-in Python objects
sites = ["Pine Creek", "Maple Wetland", "Cedar Marsh", "Oak Ridge"]
ph = [6.8, 7.2, 6.5, 5.9]
dissolved_oxygen_mg_l = [8.1, 7.4, 6.8, 9.0]

sites
````

````python
# Pair each site with its pH value
for site, value in zip(sites, ph):
    print(site, "pH =", value)
````

## 9. First look at a table with pandas

`pandas` is a major Python package for tabular data. We will use it often later. For now, just observe that a notebook can display tables directly below code cells.


````python
import pandas as pd

data = pd.DataFrame({
    "site": sites,
    "pH": ph,
    "dissolved_oxygen_mg_l": dissolved_oxygen_mg_l,
})

data
````

````python
data.describe()
````

## 10. First plot

Notebooks are especially useful when you are exploring data visually. Run the next cell to create a simple plot.


````python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(data["site"], data["pH"])
plt.axhline(7.0, linestyle="--")
plt.ylabel("pH")
plt.title("pH by monitoring site")
plt.xticks(rotation=30, ha="right")
plt.show()
````

## 11. VS Code features that help with notebooks

VS Code adds several workflow tools around notebooks. You do not need to master them today, but you should know they exist.

### Features to notice during this lesson

- **Kernel picker**: choose the Python environment that runs the notebook.
- **Variable view / variable explorer**: inspect objects currently stored in the kernel.
- **Data Viewer**: open table-like variables for filtering and inspection.
- **Autocomplete / IntelliSense**: get suggestions while writing code.
- **Linting**: get warnings about possible code problems.
- **Formatting**: make code more readable without changing what it does.
- **Debugging**: step through code when simple output inspection is not enough.
- **Outline / headings**: use Markdown headings to navigate longer notebooks.
- **Run All / Restart / Clear Outputs**: manage reproducibility and notebook state.
- **Version control with Git**: save meaningful milestones as your notebook evolves.

## 12. Debugging mindset: slow down the code

Debugging means inspecting code while it runs. In VS Code notebooks, you can debug cells when you need more than printed output.

For today, use a simpler habit first: add small print statements and inspect intermediate results.


````python
# Intentional workflow example: inspect intermediate values
mean_ph = sum(ph) / len(ph)
print("Number of pH values:", len(ph))
print("Mean pH:", mean_ph)

for site, value in zip(sites, ph):
    difference = value - mean_ph
    print(site, "difference from mean pH:", round(difference, 2))
````

## 13. Common notebook problems and fixes

| Symptom | Likely cause | First thing to try |
|---|---|---|
| `NameError` | A variable has not been created in this kernel session | Run earlier cells or restart and run all |
| `ModuleNotFoundError` | The package is not installed in the selected environment | Check the kernel and install into that environment |
| Output seems outdated | The code changed but the cell was not re-run | Re-run the cell |
| Notebook works for me but not someone else | Hidden state, missing data, or different environment | Restart and run all; check paths and environment |
| Git diff is noisy | Notebook outputs changed | Clear outputs before committing when appropriate |
