## 2. Scalar variables and data types

A **variable** stores a value.

A **scalar variable** stores a single value, such as one number, one text string, or one True/False value.

Python values have **data types**. A data type tells Python what kind of value something is.

### Integers

Integers are whole numbers.

````python
observation_count = 15
print(observation_count)
print(type(observation_count))
````

### Floating-point numbers

Floating-point numbers, or floats, are numbers with decimals.

````python
rainfall_mm = 12.7
print(rainfall_mm)
print(type(rainfall_mm))
````

### Strings

Strings store text.

````python
site_name = "Duke Forest"
print(site_name)
print(type(site_name))
````

### Booleans

Booleans store `True` or `False` values.

````python
is_raining = True
print(is_raining)
print(type(is_raining))
````

### Variable naming guidelines

Good variable names make code easier to understand.

Recommended:

```python
mean_temperature_c = 22.4
site_elevation_m = 120
```

Avoid vague names:

```python

```

Use names that describe the data and, when useful, include units.

## 3. Collections

Collections store multiple values.

We will introduce four important Python collections:

- Tuples
- Lists
- Dictionaries
- Sets

### Tuples

A tuple is an ordered collection that cannot be changed after it is created.

Tuples are useful for values that belong together and should not be modified, such as coordinates.

````python
coordinates = (35.9940, -78.8986)
print(coordinates)
print(type(coordinates))
````

Use square brackets to access values by position.

Python indexing starts at `0`.

````python
latitude = coordinates[0]
longitude = coordinates[1]

print(latitude)
print(longitude)
````

### Lists

A list is an ordered collection that can be modified.

Lists are useful for storing repeated measurements.

````python
temperatures_c = [18.1, 19.5, 21.2, 20.7]
print(temperatures_c)
print(type(temperatures_c))
````

Access a value by index:

````python
print(temperatures_c[2])
````

Add a value to a list:

````python
temperatures_c.append(22.4)
print(temperatures_c)
````

Find the number of values in a list:

````python
print(len(temperatures_c))
````

### Dictionaries

A dictionary stores **key-value pairs**.

Dictionaries are useful when values have labels or attributes.

````python
site = {
    "name": "Duke Forest",
    "state": "NC",
    "elevation_m": 120,
    "active": True
}

print(site)
print(type(site))
````

Access values using keys:

````python
print(site["name"])
print(site["elevation_m"])
````

Add a new key-value pair:

````python
site["land_cover"] = "forest"
print(site)
````

### Sets

A set stores unique values.

Sets are useful when you care about membership or unique categories.

````python
land_cover_classes = {"forest", "wetland", "urban", "forest", "agriculture"}
print(land_cover_classes)
print(type(land_cover_classes))
````

Notice that duplicate values are removed.

## 4. Conditional statements

Conditional statements allow code to make decisions.

A basic conditional asks: **If this condition is true, what should happen?**

### `if`

````python
temperature_c = 35

if temperature_c > 30:
    print("Hot day")
````

The indented line runs only if the condition is true.

### `if` and `else`

````python
temperature_c = 18

if temperature_c > 30:
    print("Hot day")
else:
    print("Not a hot day")
````

### `if`, `elif`, and `else`

````python
temperature_c = 25

if temperature_c < 10:
    print("Cold")
elif temperature_c < 25:
    print("Mild")
else:
    print("Warm")
````

## 5. Iteration with loops

Loops repeat actions.

Loops are especially important in data analysis because they allow us to process values one at a time.

### `for` loops

A `for` loop iterates over a collection.

````python
temperatures_c = [18, 20, 21, 19]

for temp in temperatures_c:
    print(temp)
````

### Example: Calculating a mean with a loop

````python
temperatures_c = [18, 20, 21, 19]

total = 0

for temp in temperatures_c:
    total = total + temp

mean_temperature = total / len(temperatures_c)
print(mean_temperature)
````

This is not the only way to calculate a mean in Python, but it helps show how loops process values step by step.

### `range()`

`range()` creates a sequence of numbers.

````python
for i in range(5):
    print(i)
````

The sequence starts at 0 and stops before 5.

You can also specify a start and stop value:

````python
for i in range(1, 6):
    print(i)
````

### `while` loops

A `while` loop repeats as long as a condition is true.

````python
count = 0

while count < 5:
    print(count)
    count = count + 1
````

Be careful with `while` loops. If the condition never becomes false, the loop can run forever.

### `break`

`break` stops a loop early.

````python
for number in range(10):
    if number == 5:
        break
    print(number)
````

### `continue`

`continue` skips the current iteration and moves to the next one.

````python
for number in range(6):
    if number == 3:
        continue
    print(number)
````
