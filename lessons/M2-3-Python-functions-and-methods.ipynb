## 1. Built-in Functions

Python includes many **built-in functions** that are available as soon as Python starts. You have probably already used several of them.

A function is a named operation that can receive input, perform some work, and often return output.

The general structure is:

```python
function_name(parameter_1, parameter_2, ...)
```

Key terms:

- **Function name**: the name used to call the function, such as `print`, `len`, or `type`.
- **Parameters**: the named inputs a function is designed to accept.
- **Arguments**: the actual values you provide when calling the function.
- **Output / return value**: the value produced by the function.

In casual conversation, people often use *parameter* and *argument* interchangeably, but technically they are different.

````python
# A few common built-in functions
site_name = "Blackwater River"
measurements = [12.4, 13.1, 11.9, 14.2]

print(site_name)
print(len(site_name))
print(len(measurements))
print(type(measurements))
print(max(measurements))
print(min(measurements))
print(sum(measurements))
````

### Function names, parameters, and outputs

Consider this expression:

```python
round(13.6789, 2)
```

Here:

- `round` is the function name
- `13.6789` is the number we want to round
- `2` is the number of decimal places
- the output is `13.68`

````python
rounded_value = round(13.6789, 2)
rounded_value
````

A function's output can be stored in a variable, printed, or used directly in another expression.

````python
temperature_c = 21.6789
rounded_temperature = round(temperature_c, 1)

print("Original temperature:", temperature_c)
print("Rounded temperature:", rounded_temperature)
````

### Some functions return values; some primarily display information

`print()` displays information to the screen. Other functions, such as `len()`, `sum()`, or `round()`, return values that are often stored or reused.

````python
# len() returns a value that we can store
number_of_measurements = len(measurements)

# print() displays information
print("Number of measurements:", number_of_measurements)
````

## 2. Keyword Arguments

Many functions allow arguments to be provided by position or by name.

For example, `round(13.6789, 2)` uses positional arguments. Python understands that the first argument is the number and the second argument is the number of digits.

Some functions also accept **keyword arguments**, where you explicitly provide the parameter name.

````python
# print() has optional keyword arguments
print("Blackwater", "River", sep="-")
print("Blackwater", "River", sep=" | ")
````

Keyword arguments are useful because they make code more readable, especially when a function has many optional settings.

## 3. Python Objects Have Methods

In Python, nearly everything is an **object**. Strings, lists, dictionaries, sets, numbers, and many other things are objects.

Objects can store data, and they can also have their own built-in operations. These object-specific operations are called **methods**.

The general structure of a method call is:

```python
object.method_name(parameter_1, parameter_2, ...)
```

A method looks similar to a function, but it is called using dot notation after an object.

````python
site_name = "blackwater river"

# String methods
print(site_name.upper())
print(site_name.title())
print(site_name.replace("river", "watershed"))
````

### Methods depend on object type

Different object types have different methods.

For example, strings have methods like `.upper()` and `.replace()`. Lists have methods like `.append()` and `.sort()`.

````python
measurements = [12.4, 13.1, 11.9]

measurements.append(14.2)
print(measurements)

measurements.sort()
print(measurements)
````

### Functions vs. methods

Functions and methods both perform operations, but they are called differently.

```python
len(site_name)          # function
site_name.upper()       # method
```

A useful way to think about the difference:

- A **function** is called by name and receives objects as input.
- A **method** belongs to a specific object and is called from that object.

The distinction matters because methods are tied to the object's type.

````python
site_name = "blackwater river"
values = [10, 20, 30]

print(len(site_name))    # len() works on strings
print(len(values))       # len() also works on lists

print(site_name.upper()) # upper() is a string method
# print(values.upper())  # This would cause an error because lists do not have an upper() method
````

### Methods that return new values vs. methods that modify objects

Some methods return a new value without changing the original object. String methods usually behave this way.

Some methods modify the object directly. List methods often behave this way.

````python
# String methods usually return a new string
site_name = "blackwater river"
new_site_name = site_name.title()

print("Original:", site_name)
print("New:", new_site_name)
````

````python
# List methods may modify the list directly
values = [3, 1, 2]
values.sort()

print(values)
````

This distinction is very important in data analysis. Some operations create a new object, while others modify an existing object.

## 5. Creating Your Own Functions

Built-in functions and object methods are powerful, but you will often want to create your own functions.

A custom function is useful when you have a task that you want to reuse, test, or make easier to read.

The general structure is:

```python
def function_name(parameter_1, parameter_2):
    # code block
    return output
```

Important syntax rules:

- Use the `def` keyword.
- Use a function name followed by parentheses.
- Put parameters inside the parentheses.
- End the first line with a colon.
- Indent the function body.
- Use `return` to send a value back from the function.

### A function with one input

This function converts temperature from Celsius to Fahrenheit.

````python
def celsius_to_fahrenheit(temp_c):
    temp_f = temp_c * 9 / 5 + 32
    return temp_f
````

````python
air_temp_c = 22.5
air_temp_f = celsius_to_fahrenheit(air_temp_c)

print(air_temp_f)
````

Once a function is defined, it can be reused with different inputs.

````python
print(celsius_to_fahrenheit(0))
print(celsius_to_fahrenheit(20))
print(celsius_to_fahrenheit(30))
````

### A function with two inputs

Functions can accept more than one input.

This function calculates the mean of two measurements.

````python
def mean_of_two(value_1, value_2):
    mean_value = (value_1 + value_2) / 2
    return mean_value
````

````python
mean_stream_temp = mean_of_two(17.4, 18.2)
print(mean_stream_temp)
````

### A function with two outputs

A function can return more than one value. Python does this by returning multiple values separated by commas.

This function returns both the minimum and maximum value from a list.

````python
def min_and_max(values):
    minimum_value = min(values)
    maximum_value = max(values)
    return minimum_value, maximum_value
````

````python
stream_temps_c = [17.4, 18.2, 16.9, 19.1, 18.7]

low_temp, high_temp = min_and_max(stream_temps_c)

print("Lowest temperature:", low_temp)
print("Highest temperature:", high_temp)
````

Returning two values is common when a function calculates related pieces of information.

````python
def summarize_values(values):
    count = len(values)
    mean_value = sum(values) / count
    return count, mean_value

n_obs, mean_temp = summarize_values(stream_temps_c)

print("Number of observations:", n_obs)
print("Mean temperature:", mean_temp)
````
