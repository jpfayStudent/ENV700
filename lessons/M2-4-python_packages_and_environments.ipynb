## 1. Why packages matter

Python gives us the core language: variables, lists, dictionaries, loops, functions, and other building blocks. Packages extend Python by adding specialized tools.

For environmental data analysis, we often use packages for:

- reading and organizing tabular data,
- working with file paths,
- calculating summary statistics,
- creating plots,
- analyzing spatial or time-series data.

Instead of writing everything from scratch, we import packages written and tested by others.

## 2. Common import patterns

You will see three common import patterns in this course.

| Pattern | Example | What it means |
|---|---|---|
| Import the whole package | `import math` | Use the package name before its tools, such as `math.sqrt()` |
| Import with an alias | `import pandas as pd` | Use a shorter conventional nickname, such as `pd.DataFrame()` |
| Import one tool from a package | `from pathlib import Path` | Use that tool directly, such as `Path("data")` |

Aliases are common when a package is used frequently. For example, `pandas` is almost always imported as `pd`.

````python
import math

math.sqrt(25)
````

````python
import pandas as pd

pd.DataFrame({
    "site": ["A", "B", "C"],
    "temperature_c": [18.2, 21.5, 19.8]
})
````

````python
from pathlib import Path

project_folder = Path(".")
project_folder
````

## 3. Using tools from packages

After a package is imported, you can use the functions, objects, and methods it provides.

A useful pattern is:

```python
package_name.tool_name()
```

For example, after running `import math`, the square root function is available as `math.sqrt()`.

````python
import math

value = 49
root = math.sqrt(value)

root
````

### A small pandas example

`pandas` is a third-party package for working with tabular data. A `DataFrame` is the main pandas object for storing rows and columns.

````python
import pandas as pd

data = pd.DataFrame({
    "lake": ["Mendota", "Monona", "Wingra"],
    "depth_m": [25.3, 22.5, 4.3],
    "surface_area_km2": [39.4, 13.2, 1.4]
})

data
````

## 4. Standard library vs. third-party packages

Some packages are included automatically when Python is installed. These are part of Python's **standard library**.

Other packages are created outside of core Python and must be installed separately. These are called **third-party packages**.

| Standard library examples | Third-party examples |
|---|---|
| `math` | `pandas` |
| `pathlib` | `numpy` |
| `datetime` | `matplotlib` |
| `statistics` | `geopandas` |

In this course, many of our most important tools will be third-party packages.

## 5. Conda environments: the basic idea

Our course uses **Miniforge** and **Conda** to manage Python and its packages.

A Conda environment is a self-contained collection of:

- a Python version,
- installed packages,
- package versions that are designed to work together.

This matters because data analysis projects often depend on several packages at once. Environments help keep those tools organized and reproducible.

For now, the key point is simple: when you work in VS Code or Jupyter, make sure you are using the correct course environment/kernel.

### Installing packages

You will usually install packages from the terminal, not from inside the notebook.

Common Conda command:

```bash
conda install package-name
```

Example:

```bash
conda install pandas
```

Sometimes you may see `pip install package-name`. `pip` is another Python package installer. In this course, prefer Conda unless instructed otherwise.

## 7. Mini workflow: packages working together

In real analyses, packages are often used together. Here we use pandas to create a small dataset and pandas' built-in plotting method to make a quick exploratory plot.

````python
import pandas as pd

rainfall = pd.DataFrame({
    "year": [2021, 2022, 2023, 2024],
    "rainfall_mm": [940, 1015, 875, 960]
})

rainfall
````

````python
rainfall.plot(x="year", y="rainfall_mm", kind="line")
````

The point is not the plot itself. The point is that packages give us reusable tools for common data tasks.
